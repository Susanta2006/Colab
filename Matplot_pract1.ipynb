Importing libraries, then collecting data, and coeffs represents here coefficients and y_line uses formula y=mx+c

```python
import matplotlib.pyplot as plt
import numpy as np

# sample dataset
x_data=np.array([1,2,3,4,5,6,7,8])
y_data=np.array([2.1,3.9,6.2,7.8,10.1,12.3,14.0,15.8])

#create model line(linear fit)
coeffs=np.polyfit(x_data,y_data,1)
x_line=np.linspace(1,8,100)
y_line=coeffs[0]*x_line+coeffs[1]
```

Now lets Create the plotting line.. and see the output

```python
plt.figure(figsize=(8,6))
plt.scatter(x_data,y_data,color='red',s=80, alpha=0.7, label='Data Points')
plt.plot(x_line,y_line,color='blue',linewidth=2,label='Linear Fit')

#add labels and formatting
plt.xlabel('X-axis',fontsize=14)
plt.ylabel('Y-axis',fontsize=14)
plt.title('Linear Regression Example',fontsize=16)
plt.legend()
plt.grid(True,alpha=0.4)
plt.show()
```
