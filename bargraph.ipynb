Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

calling data

```python
student=[1,2,3,4,5]
math=[60,70,80,90,100]
science=[55,65,78,76,85]
```

plotting the graph

```python
plt.bar(student,math,label='maths')
plt.bar(student,science,label='science')
plt.xlabel('student')
plt.ylabel('marks')
plt.legend()
plt.title('student marks')
plt.show()
```
