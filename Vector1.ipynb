```python
import numpy as np
# Define two 2D vectors
A = np.array([3, -2])
B = np.array([4, 5])

print("Vector A:", A)
print("Vector B:", B)

# 1. Vector Addition
C = A + B
print("A + B =", C)

# 2. Vector Subtraction
D = A - B
print("A - B =", D)

# 3. Scalar Multiplication
E = 2 * A
print("2 * A =", E)

# (Optional extra: Dot Product, just for completeness)
dot_product = np.dot(A, B)
print("A · B =", dot_product)
```
